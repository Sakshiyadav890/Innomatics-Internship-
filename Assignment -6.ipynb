```python
def event_registration_controller(capacity, registrations):
    """
    This function manages event registrations.
    
    Parameters:
    capacity (int): Maximum number of allowed participants
    registrations (int): Total number of people who registered
    
    Returns:
    dict: Contains confirmed registrations, waitlisted users, and status
    """
    
    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    
    return {
        "Confirmed Registrations": confirmed,
        "Waitlisted Users": waitlisted,
        "Registration Status": status
    }
```

```python
capacity = 100
registrations = 105
```

```python
result = event_registration_controller(capacity, registrations)
```

```python
print("Confirmed Registrations:", result["Confirmed Registrations"])
print("Waitlisted Users:", result["Waitlisted Users"])
print("Registration Status:", result["Registration Status"])
```

```python
def event_registration_controller_v2(capacity, registrations_list):
    """
    Advanced version using list of users
    """
    confirmed_users = []
    waitlisted_users = []
    
    for i, user in enumerate(registrations_list):
        if i < capacity:
            confirmed_users.append(user)
        else:
            waitlisted_users.append(user)
    
    status = "Closed" if len(registrations_list) > capacity else "Open"
    
    return confirmed_users, waitlisted_users, status


users = ["User" + str(i) for i in range(1, 106)]

confirmed, waitlisted, status = event_registration_controller_v2(100, users)

print("Confirmed:", len(confirmed))
print("Waitlisted:", len(waitlisted))
print("Status:", status)
```
