```python
import re
from collections import Counter
```

```python
def preprocess_text(text):
    if not text or text.strip() == "":
        return []

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove emails
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Convert to lowercase
    text = text.lower()

    # Handle repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove special characters (keep words only)
    text = re.sub(r"[^a-z\s]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenize
    tokens = text.split()

    # Remove short words (<=2 except no, not)
    filtered_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            filtered_tokens.append(word)

    return filtered_tokens
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product ",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
results = []

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    clean_sentence = " ".join(tokens)

    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", clean_sentence)
    print("-" * 50)

    results.append(tokens)
```

```python
for tokens in results:
    if len(tokens) == 0:
        continue

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total

    print("Tokens:", tokens)
    print("Total:", total)
    print("Unique:", unique)
    print("Avg Length:", round(avg_len, 2))
    print("-" * 40)
```

```python
all_tokens = []

for tokens in results:
    all_tokens.extend(tokens)

freq = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(freq.most_common(10))

print("\nLeast 5 Frequent Words:")
print(freq.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        clean_sentences.append(" ".join(tokens))
        all_tokens.extend(tokens)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
print(preprocess_text(""))          
print(preprocess_text("12345"))     
print(preprocess_text(""))    
```
