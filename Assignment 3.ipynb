## Problem 1: Employee Performance Bonus Eligibility
This program identifies top-performing employees eligible for bonus.

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
max_score = max(employees.values())

# Find employees with highest score
top_performers = []

for name, score in employees.items():
    if score == max_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), "(Score:", max_score, ")")
```

## Problem 2: Search Query Keyword Analysis
This program counts frequency of keywords.

```python
text = "Buy mobile phone buy phone online"

# Convert to lowercase
text = text.lower()

# Remove punctuation (basic)
words = text.split()

freq = {}

for word in words:
    freq[word] = freq.get(word, 0) + 1

# Print words with frequency > 1
result = {}

for key, value in freq.items():
    if value > 1:
        result[key] = value

print(result)
```

## Problem 3: Sensor Data Validation
This program filters valid (even) sensor readings.

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        valid_readings.append((i, sensor_readings[i]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## Problem 4: Email Domain Usage Analysis
This program calculates domain usage percentage.

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(domain + ":", str(int(percentage)) + "%")
```

## Problem 5: Sales Spike Detection
This program detects unusually high sales.

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
avg = sum(sales) / len(sales)

# Detect spikes (>30% above average)
threshold = avg * 1.3

for i in range(len(sales)):
    if sales[i] > threshold:
        print("Day", i+1, ":", sales[i])
```

## Problem 6: Duplicate User ID Detection
This program finds duplicate user IDs and their count.

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count = {}

for user in user_ids:
    count[user] = count.get(user, 0) + 1

for user, freq in count.items():
    if freq > 1:
        print(user, "→", freq, "times")
```
