## Problem 1: Social Media – Post Engagement Analyzer
This function calculates total likes and determines whether a post is viral or not.

```python
def post_engagement(likes):
    total_likes = 0
    
    # Loop to calculate total likes
    for like in likes:
        total_likes += like

    # Condition to check viral status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

# Example
post_engagement([200, 300, 250, 300])
```

## Problem 3: Rainfall Adequacy Checker
This function calculates average rainfall and checks adequacy.

```python
def rainfall_checker(data, required):
    total = 0

    for rain in data:
        total += rain

    avg = total / len(data)

    if avg >= required:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", avg)
    print("Rainfall Status:", status)

rainfall_checker([70, 80, 65, 75], 70)
```

## Problem 4: Duplicate Account Detection
This function detects duplicate usernames.

```python
def check_duplicates(users):
    if len(users) != len(set(users)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

check_duplicates(["user1", "user2", "user3", "user1"])
```

## Problem 6: Premium Crop Price Filter
This function filters premium crops based on price.

```python
def premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)

premium_crops([1500, 2500, 3200, 1800])
```

## Problem 7: Application Health Checker
This function checks system health based on errors.

```python
def app_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

app_health(7)
```

## Problem 9: Attendance Eligibility System
This function calculates attendance and eligibility.

```python
def attendance_checker(attended, total):
    percentage = (attended / total) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance_checker(40, 50)
```

## Problem 10: Smart Electricity Bill Analyzer
This function calculates bill and classifies usage.

```python
def electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Level:", usage)

electricity_bill(250)
```
