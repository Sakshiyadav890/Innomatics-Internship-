## Problem 1: Mobile Recharge Validation System
This function validates recharge amount and plan with retry option.

```python
def recharge_validation():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        if amount < 50:
            print("Invalid: Minimum recharge is ₹50")
        elif amount not in valid_plans:
            print("Invalid plan. Try again.")
        else:
            print("Recharge Successful!")
            break

recharge_validation()
```

## Problem 2: Inventory Reorder Alert System
This function checks stock levels of products.

```python
def inventory_alert(products):
    for item, stock in products.items():
        if stock < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")

products = {
    "Rice": 10,
    "Wheat": 20,
    "Sugar": 12
}

inventory_alert(products)
```

## Problem 3: Student Result Processing System
This function calculates average marks and determines pass/fail.

```python
def student_result(marks):
    total = 0

    for m in marks:
        total += m

    avg = total / len(marks)

    if avg >= 50:
        result = "Pass"
    else:
        result = "Fail"

    print("Average Marks:", avg)
    print("Result:", result)

student_result([60, 70, 50, 40])
```

## Problem 4: Cab Fare Estimator
This function calculates cab fare with retry option.

```python
def cab_fare():
    while True:
        km = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")

        fare = 50 + (km * 12)

        if peak.lower() == "yes":
            fare += fare * 0.25

        print("Total Fare:", fare)

        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            break

cab_fare()
```

## Problem 5: Employee Attendance Eligibility Checker
This function calculates attendance percentage.

```python
def attendance_checker(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance %:", percentage)
    print("Status:", status)

attendance_checker(["P", "P", "A", "P", "P", "A", "P"])
```

## Problem 6: Password Strength Checker
This function checks password strength.

```python
def password_checker(password):
    has_digit = False
    has_special = False

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if len(password) >= 8 and has_digit and has_special:
        print("Strong Password")
    else:
        print("Weak Password")

password_checker("Pass@123")
```
